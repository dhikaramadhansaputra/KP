```python
# Path direktori tempat file Excel berada
data_dir = r"D:\Dhika Ramadhan Saputra\Data Analist\Full Project\Tren Penerimaan MABA UMP 2014 - 2024\dataset\raw"

# List nama file
file_names = [f"(raw) {year}.xlsx" for year in range(2014, 2025)]

# List untuk menyimpan dataframe dari setiap file
dfs = []

# Loop melalui setiap file, baca sebagai dataframe, dan tambahkan ke list
for file in file_names:
    try:
        file_path = f"{data_dir}\{file}"  # Gabungkan path direktori dengan nama file
        df = pd.read_excel(file_path, skiprows=4)  # Lewati 4 baris pertama
        dfs.append(df)
    except FileNotFoundError:
        print(f"File {file} tidak ditemukan di {data_dir}. Melewati...")

# Gabungkan semua dataframe dalam list menjadi satu dataframe
df_gabungan = pd.concat(dfs, ignore_index=True)

# Tampilkan info dataframe gabungan
print(df_gabungan.info())

# Tampilkan beberapa baris pertama dataframe gabungan
df_gabungan
```
